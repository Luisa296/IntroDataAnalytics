## Pregunta 6
**Agrupar por Retailer y calcular:**
- La suma de 'Total Sales'
- La suma de 'Operating Profit'
- La suma de 'Units Sold'

_Utiliza `groupby` para obtener estos totales por cada Retailer._

```python
data.groupby('Retailer').agg({'Total Sales':"sum", 'Operating Profit':"sum", 'Units Sold':"sum"})
```

## Pregunta 7
**Calcular el margen total por Retailer y el margen operativo promedio por Retailer.**

- Margen total: (Suma de 'Operating Profit' / Suma de 'Total Sales') por Retailer.
- Margen promedio: Promedio de la columna 'Operating Margin' para cada Retailer.

_Compara ambos resultados y comenta la diferencia, reflexionando sobre el significado en términos de negocio._

El margen total y el margen operativo promedio ofrecen perspectivas diferentes sobre la rentabilidad de un retailer. Mientras que el margen total muestra la ganancia global, el margen promedio indica la rentabilidad típica de una venta individual. Ambas medidas son utiles y al ser comparadas le brinda información a las empresas para identificar áreas de mejora, ya sea en volumen de ventas, estrategia de precios o eficiencia operativa, para optimizar su desempeño.

```python
Retailer=data.groupby('Retailer').agg({'Total Sales':"sum", 'Operating Profit':"sum"})
```
